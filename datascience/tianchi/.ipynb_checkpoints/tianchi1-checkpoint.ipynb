### 合并当前数据集，剔除元素完全一样的变量

```python
# 删除df_all里面所有值都相同的列
def dropSameCol(df):
    result_list = []
    for col in df.columns:
        count_values = df[col].value_counts()
        if count_values.shape[0] == 1:
            result_list.append(False)
        else:
            if count_values.shape[0] <= 5 and max(count_values) >= 4.0*sum(count_values)/5:
                result_list.append(False)
            else:
                result_list.append(True)
    return df.loc[:, result_list]
```

```python
# standarzation
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
y_train_std = stdsc.fit_transform(np.array(y_train).reshape(-1,1))[:,0]
y_test_std= stdsc.transform(np.array(y_test).reshape(-1,1))[:,0]
X_std = stdsc.fit_transform(X)
y_std = stdsc.fit_transform(np.array(y).reshape(-1,1))[:0]
```

## 8. 特征选取

### Filter特征选择
#### CHi2方法(只支持离散因变量，不支持连续因变量）

### 基于正则项的特征选择
#### L1 正则项

```python
def filterByCorr(x, y):
    corr_list = []
    for i in range(x.shape[1]):
        corr = np.corrcoef(x[:, i].reshape(1, -1), y)
        corr_list.append(corr[0,1])
    return (np.array(corr_list) > 0.2) | (np.array(corr_list) < -0.2)

filtered = filterByCorr(X, y)
print(Counter(filtered))
X[:, filtered].shape
```

```python
X_train_std_selected = X_train_std[:, filtered]
X_test_std_selected = X_test_std[:, filtered]
```

```python
from sklearn import linear_model

#reg = linear_model.LinearRegression()
#reg = linear_model.Ridge(alpha=0.1)
reg = linear_model.Lasso(alpha=0.01, max_iter=1e6, tol=0.001)
reg.fit(X_train_std_selected, y_train_std)
coef = reg.coef_
# print(coef)
print(reg.score(X_train_std_selected, y_train_std))
print(reg.score(X_test_std_selected, y_test_std))
```

```python
from collections import Counter
Counter(coef!=0)

X_train_std_selected2 = X_train_std_selected[:, np.array(coef) != 0]
X_test_std_selected2 = X_test_std_selected[:, np.array(coef) != 0]
```

```python
lr = linear_model.Lasso(alpha=0.001)
lr.fit(X_train_std_selected2, y_train_std)
print(lr.score(X_train_std_selected2, y_train_std))
print(lr.score(X_test_std_selected2, y_test_std))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor( max_depth=5)
dtr.fit(X_train_std_selected2, y_train_std)
print(dtr.score(X_test_std_selected2, y_test_std))
print(dtr.score(X_train_std_selected2, y_train_std))
```

## 递归特征清除

recursive feature elimination (RFE) 递归特征清除：给特征一定的权重，然后通过训练模型，发现权重较小的特征，删除，重复该流程，直到达到我们需要的特征数量（比如回归模型）

```python
# 此时需要对样本进行标准化，而非归一化
# 变量多时千万别这么干，跑一遍太慢了
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

svc = SVC(kernel="linear", C=1)
rfe = RFE(estimator=linear_model.LinearRegression(), 
          n_features_to_select=10,  # 要选出几个 feature
          step=1)  # 每次剔除出几个feature
rfe.fit(X_train_std_selected2, y_train_std)

X_train_std_selected3 = rfe.transform(X_train_std_selected2)
X_test_std_selected3 = rfe.transform(X_test_std_selected2)
```

```python
# 查看选出了哪几个 feature
mask = rfe.get_support()
print(mask)
plt.matshow(mask.reshape(1, -1), cmap='gray_r');
```

```python
lr = linear_model.Lasso(alpha=0.001)
lr.fit(X_train_std_selected3, y_train_std)
print(lr.score(X_train_std_selected3, y_train_std))
print(lr.score(X_test_std_selected3, y_test_std))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor( max_depth=4, min_samples_split=10)
dtr.fit(X_train_std_selected3, y_train_std)
print(dtr.score(X_test_std_selected3, y_test_std))
print(dtr.score(X_train_std_selected3, y_train_std))
```
